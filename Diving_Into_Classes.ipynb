```python
import pandas as pd 
import os 
import sys 
import shutil 
```

```python
labels = pd.read_csv(r'D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\train.csv')
labels
```

```python
train_dir = r'D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\train_images'
DIR = r"D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\Class"
TEST = r"D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\TEST"
TRAIN = r"D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\TRAIN"
VALIDATION = r"D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\VALIDATION"
Class = r"D:\Soumyadeep Das\CODING\Surajit Sir\cassava-leaf-disease-classification\Class"
```

```python
# for creating subclasses

if not os.path.exists(DIR):
    os.mkdir(DIR)

for image_id,label in labels.values:
    if not os.path.exists(DIR+str(label)):
        os.mkdir(DIR+str(label))
    src_path = train_dir+'/'+image_id
    dst_path = DIR+str(label)+'/'+image_id
    shutil.copy(src_path,dst_path)
```

```python
total = 0
for folder in os.listdir(Class):

    # initialization of the variables 
    count = 0
    test_value = 0
    train_value = 0
    validation_value = 0
    extra = 0
    
    sub_class = Class+'/'+folder
    print(folder)

    # for getting the Count of number of images in every classes
    for i in os.listdir(sub_class):
        count = count+1

    # some variables for better understanding of every classes 
    total = total+count
    test_value = int(count*0.1)
    train_value  = int(count*0.8)
    validation_value = int(count*0.1)
    extra = (test_value+train_value+validation_value)-count

    if extra>0:
        extra = extra 
    else:
        extra = -extra 

    DIR1 = TRAIN+'/'+folder
    DIR2 = TEST+'/'+folder
    DIR3 = VALIDATION+'/'+folder
    
    # for copying Train images from every classes 
    for i in range (1,train_value+1):

        if not os.path.exists(DIR1):
            os.mkdir(DIR1)
        
        src_path = os.listdir(sub_class)[i]
        src_path = sub_class+'/'+src_path
        dst_path = DIR+'/'+os.listdir(sub_class)[i]
        
        shutil.copy(src_path,dst_path)

    # for copying Test images from every classes 
    for i in range (train_value,train_value+test_value):

        if not os.path.exists(DIR2):
            os.mkdir(DIR2)
        
        src_path = os.listdir(sub_class)[i]
        src_path = sub_class+'/'+src_path
        dst_path = DIR2+'/'+os.listdir(sub_class)[i]
        
        shutil.copy(src_path,dst_path)

    # for copying Validation images from every classes
    for i in range (train_value+test_value,train_value+test_value+validation_value):

        if not os.path.exists(DIR3):
            os.mkdir(DIR3)
        
        src_path = os.listdir(sub_class)[i]
        src_path = sub_class+'/'+src_path
        dst_path = DIR3+'/'+os.listdir(sub_class)[i]
        
        shutil.copy(src_path,dst_path)

    
    print("Count->",count)
    print("Test->",test_value)
    print("Train->",train_value)
    print("Validation->",validation_value)
    print("TTV->",test_value+train_value+validation_value)
    print("Extra->",extra)

        

print("Total Data->",total)
```
